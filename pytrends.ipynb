```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pytrends.request import TrendReq
```

```python
pytrends = TrendReq(hl='en-US', tz=-330, geo='IN')
```

```python
timeframe_dict = {
    'year_start': 2018,
    'month_start': 1,
    'day_start': 1,
    'hour_start': 0,
    'year_end': 2019, 
    'month_end': 10,
    'day_end': 31,
    'hour_end': 0
}
```

```python
kw_list = ["Sharad Pawar", 
           "Devendra Fadanvis", 
           "Aditya Thackeray",
           "Sanjay Raut"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 1-m', gprop='')
df = pytrends.interest_over_time()
df[kw_list].plot(figsize=(15, 8))
```
